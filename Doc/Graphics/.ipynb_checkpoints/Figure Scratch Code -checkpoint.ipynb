```python
def get_features(dirt, saveFig=False):
    dig_count = dirt.sum
    
    # Compute the medial axis (skeleton) and the distance transform
    skel, distance = medial_axis(dirt, return_distance=True)
    contours = measure.find_contours(dirt, 0)
    
    dist_contours = distance
    dist_skel_contours = distance * skel
    
    contours_length = 0
    max_dist = np.max(dist_skel_contours)
    for n, contour in enumerate(contours):
        contours_length += contour.shape[0]
        for point in contour:
            x = int(point[0])
            y = int(point[1])
            dist_contours[x, y] = max_dist
            dist_skel_contours[x, y] = max_dist
    
    skel_length = skel.sum()
    print "skel_length: ", skel_length
    print "contours_length: ", contours_length
    
    
    if(saveFig):
        plt.imsave('dist_contours.png', dist_contours, cmap=plt.cm.spectral)
        plt.imsave('dist_skel_contours.png', dist_skel_contours, cmap=plt.cm.spectral)
    return dist_skel_contours
```

```python
import PIL
from PIL import Image
from matplotlib import pyplot as plt
import colorsys



im = Image.open('dist_contours.png')  
w, h = im.size  
colors = im.getcolors(w*h)
colors = colors[:250]

def hexencode(rgb):
    r=rgb[0]
    g=rgb[1]
    b=rgb[2]
    return '#%02x%02x%02x' % (r,g,b)

def get_hsv(rgb):
    
    r=rgb[0]
    g=rgb[1]
    b=rgb[2]
    hexrgb = '#%02x%02x%02x' % (r,g,b)
    
    hexrgb = hexrgb.lstrip("#")   # in case you have Web color specs
    r, g, b = (int(hexrgb[i:i+2], 16) / 255.0 for i in xrange(0,5,2))
    return colorsys.rgb_to_hsv(r, g, b)

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)

colors.sort(key=lambda x: x[0])

ax.set_title('Distance Histogram Plot')
ax.set_ylabel('Frequency')
ax.set_xlabel('Sorted Distances by Frequency')
ax.set_xticklabels('')

for idx, c in enumerate(colors):
    ax.bar(idx, c[0], color=hexencode(c[1]),edgecolor=hexencode(c[1]))

ax.grid(b=True,axis='y')
ax.set_yscale('linear')
plt.savefig('dist_count_linear.svg')
plt.savefig('dist_count_linear.png')
ax.set_yscale('symlog')
plt.savefig('dist_count_symlog.svg')
plt.savefig('dist_count_symlog.png')
plt.show()
```
