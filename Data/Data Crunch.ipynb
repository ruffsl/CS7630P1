```python
def plot_group(group, mode):    
    mean_stat = np.asarray(group.mean(axis=1))
#     mean_stat = np.asarray(group.median(axis=1))
    max_stat = np.asarray(group.max(axis=1))
    min_stat = np.asarray(group.min(axis=1))
    sdf_stat = np.asarray(group.std(axis=1))
    std_p_stat = np.asarray(mean_stat + sdf_stat)
    std_n_stat = np.asarray(mean_stat - sdf_stat)
#     std_p_stat = np.percentile(group.as_matrix(),75, axis=1)
#     std_n_stat = np.percentile(group.as_matrix(),25, axis=1)
    
    
    x = np.arange(max_stat.shape[0])
    
    fig, ax = plt.subplots(figsize=(8,6))
    
    ax.fill_between(x, max_stat, min_stat,
                    alpha=0.5, color='blue', linewidth = 3,
                    label='Max and Min')
    max_min_prox, = ax.plot([0],
                    alpha=0.5, color='blue', linewidth = 3,
                    label='Max and Min')
    
    ax.fill_between(x, std_p_stat, std_n_stat,
                    alpha=0.6, color='green', linewidth = 3,
                    label='STD')
    std_prox, = ax.plot([0],
                    alpha=0.6, color='green', linewidth = 3,
                    label='STD')
    
    ax.plot(mean_stat, color='orange',
                    label='Mean', linewidth = 3)
    # ax.legend(loc='best')
    ax.set_ylabel('Dirt Moved')
    ax.set_xlabel('Iterations')
    ax.set_xbound(0,mean_stat.size)
    ax.set_ybound(0,None)
    ax.grid()
    
    leg = ax.legend(loc=2, fancybox=True)
    leg.get_frame().set_alpha(0.7)    
    max_min_prox.set_visible(False)
    std_prox.set_visible(False)
    
    figname = mode + '_' + group.keys()[0]
    ax.set_rasterized(True)
    
    plt.savefig(figname + '.eps')
    plt.savefig(figname + '.svg')
    

def plot_mode(mode):
    files = [ f for f in listdir(mode) if isfile(join(mode,f)) ]
    data = []
    for f in files:
        df = pd.read_csv(mode + '/' + f, index_col='it')
        data.append(df)
    df = pd.concat(data, axis=1, join='inner')
    grouped = df.groupby(lambda x: x[1], axis=1)
    for i, group in grouped:
        fig = plot_group(group, mode)
        
```
